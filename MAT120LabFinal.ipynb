### **Section 29, 30: Question 1**
The Factorial of a number is defined as $$ n! = \int_{0}^{\infty} x^n e^{-x}dx = \Gamma(n+1)$$
  Now plot $x!$ vs x for $x\in[0,1]$ with step size, $h=0.01$. \
   Also using trapezoidal method, evaluate $\int_{0}^{1} x!dx$ with the same step size.
   \
   ** You cannot use SymPy, SciPy, or NumPy. You can however use math and Matplotlib. ** \

  **Hints:** from math import gamma

```python
from math import gamma
from matplotlib.pyplot import *
h = 0.01
x = [i*h for i in range(101)]
y = [gamma(i+1) for i in x]
plot(x,y)
grid()
```

```python
#Area under the curve
A= 0
for i in range(101):
  if (i ==0) or (i==100):
    A = A + h* y[i]/2
  else:
    A = A + h*y[i]
print(f"Integration result using trapezoidal rule = {A}")
```

### **Section 4 Question 2**
Population of North Korea is given by the following Logistic Growth Equation: $$\frac{dP}{dt}= rP(1-\frac{P}{K})$$
Let's say for North Korea, growth rate $r=0.01$ and population capacity $K=30M$ and in 1955, the population was $P(0)= 15M$, (M= Million). Here t is measured on an annual basis. \
Solve the differential equation using Euler's method with step size $h = 0.001$ and plot the Population for the years 1955 to 2054. \

** You cannot use SymPy or SciPy. You should use Numpy and Matplotlib **


```python
from numpy import *
h = 0.001
t = linspace(0, 100, 100001) # 2054-1955+1=100, 100/h + 1= 100001
P = 0*t
r = 0.01
P[0] = 15 #M
K = 30 #M
f = lambda x, y : r*y*(1-y/K)

for i in range(1, 100001):
  P[i] = P[i-1] + h*f(t[i-1], P[i-1])

plot(t,P)
#x axis
years = [1955, 1988, 2021, 2054]
xticks(list(linspace(0,100,4)),years)

grid()
xlabel('years')
ylabel('Population in Million')
show()
```

### **Section 6, Question 1**
  The Factorial of a number is defined as $$ n! = \int_{0}^{\infty} x^n e^{-x}dx $$
  Using SymPy calculate the following accurate up to $4$ decimal points.
  $$ 		\left. \frac{d^2}{dx^2}(x!) \right|_{x=e} $$



```python
from sympy import *
x, y = symbols('x y')
fact_x = integrate(y**x * exp(-y), (y,0,oo))
diff(fact_x, x, 2).subs(x,e).evalf(5)
```

### **Section 29,30, Question 2**
Using SymPy solve the following systems of ODE and find $y(t)$:
$$\frac{dx}{dt}+ y = 0, \frac{dy}{dt}= x$$
$$ x(0)= 3, y(0)= 4$$

```python
from sympy import *
t= symbols('t')
x = Function('x')(t)
y = Function('y')(t)
eq1 = Eq(diff(x,t)+y, 0)
eq2 = Eq(diff(y,t), x)
boundary = {x.subs(t,0):3, y.subs(t,0): 4}

solution = dsolve([eq1, eq2], ics = boundary)
```
