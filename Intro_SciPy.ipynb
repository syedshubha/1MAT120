$ f(x) = xcos(2x) , x \in [0, \frac{\pi}{4}] $

```python
f = lambda x: x*cos(2*x)
x = linspace(0,pi/4, 999)
plot(x, f(x), label="f(x)")
plot(x, misc.derivative(f, x, dx=1e-6, n=1), label="f'(x)")
legend()
grid()
```

$$\int_{0}^{\pi} xcos(x/2)dx$$

```python
from sympy import symbols, integrate, cos, pi
x = symbols('x')
exact = integrate(x*cos(x/2), (x, 0, pi)).evalf()
exact
```

$$\int_{0}^{\pi} xcos(x/2)dx$$

```python
f = lambda x: x*cos(x/2)
x = linspace(0, pi, 99)
int_trap = integrate.trapz(f(x), x)
print(f"Value = {int_trap}, Error = {abs(int_trap-exact)}")
```

```python
integral, err_bound = integrate.quad(f, 0, pi)
print(f"Value = {integral}, Error = {abs(integral-exact)}")
```

$\frac{dy}{dx}= x+ y, y(0)=1$

```python
#RK Method: Runga-Kutta
dydx = lambda x, y: x+y
y0 = [1]
x = linspace(0,2,101)
sol = integrate.solve_ivp(dydx, [0,2], y0, t_eval = x)
plot(x, sol.y[0])
yticks(range(13))
grid()
show()
```

```python
#odeint
plot(x, integrate.odeint(dydx, y0, x))
yticks(range(13))
grid()
show()
```
